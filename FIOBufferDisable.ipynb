```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
fio=pd.read_csv('FIOBufferDisable.csv')
```

```python
print(fio.head())
print(fio.isnull().values.any())
```

```python
system = fio['System']
Random_Read_MBs = fio['Random Read MB/s']
Random_Read_IOPS = fio['Random Read IOPS']

Random_Write_MBs = fio['Random Write MB/s']
Random_Write_IOPS = fio['Random Write IOPS']

Sequential_Read_MBs = fio['Sequential Read MB/s']
Sequential_Read_IOPS = fio['Sequential Read IOPS']

Sequential_Write_MBs = fio['Random Write MB/s']
Sequential_Write_IOPS = fio['Random Write IOPS']

# print(Random_Write_MBs)
# print(Random_Write_IOPS)

# print(Sequential_Read_MBs)
# print(Sequential_Read_IOPS)

# print(Sequential_Write_MBs)
# print(Sequential_Write_IOPS)

fig, axes = plt.subplots(2, 4, figsize=(16, 10))
fig.suptitle('FIO Engine: Linux AIO - Buffered: No - Direct: Yes - Block Size: 4KB')

sns.boxplot(ax=axes[0, 0], x=system, y=Random_Read_MBs)
sns.boxplot(ax=axes[0, 1], x=system, y=Random_Read_IOPS)
sns.boxplot(ax=axes[0, 2], x=system, y=Random_Write_MBs)
sns.boxplot(ax=axes[0, 3], x=system, y=Random_Write_IOPS)
sns.boxplot(ax=axes[1, 0], x=system, y=Sequential_Read_MBs)
sns.boxplot(ax=axes[1, 1], x=system, y=Sequential_Read_IOPS)
sns.boxplot(ax=axes[1, 2], x=system, y=Sequential_Write_MBs)
sns.boxplot(ax=axes[1, 3], x=system, y=Sequential_Write_IOPS)
```
