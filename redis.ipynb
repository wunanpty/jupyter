```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
redis=pd.read_csv('Redis.csv')
```

```python
print(redis.head())
print(redis.isnull().values.any())
```

```python
system = redis['System']
LPOP = redis['LPOP(RPS)']
SADD = redis['SADD(RPS)']
GET = redis['GET(RPS)']
SET = redis['SET(RPS)']
LPUSH = redis['LPUSH(RPS)']


fig, axes = plt.subplots(1, 5, figsize=(16, 10))
fig.suptitle('Redis Performance')

sns.boxplot(ax=axes[0], x=system, y=LPOP)
sns.boxplot(ax=axes[1], x=system, y=SADD)
sns.boxplot(ax=axes[2], x=system, y=GET)
sns.boxplot(ax=axes[3], x=system, y=SET)
sns.boxplot(ax=axes[4], x=system, y=LPUSH)
```
