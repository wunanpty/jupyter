```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
postgre=pd.read_csv('postgre.csv')
```

```python
print(postgre.head())
print(postgre.isnull().values.any())
```

```python
system = postgre['System']
sf1_c50_ms = postgre['ScalingFactor1 Clients50 ReadWrite ms']
sf100_c50_ms = postgre['ScalingFactor100 Clients50 ReadWrite ms']
sf1_c100_ms = postgre['ScalingFactor1 Clients100 ReadWrite ms']

sf1_c50_tps = postgre['ScalingFactor1 Clients50 ReadWrite TPS']
sf100_c50_tps = postgre['ScalingFactor100 Clients50 ReadWrite TPS']
sf1_c100_tps = postgre['ScalingFactor1 Clients100 ReadWrite TPS']


fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('PostgreSQL Performance')

sns.boxplot(ax=axes[0, 0], x=system, y=sf1_c50_ms)
sns.boxplot(ax=axes[0, 1], x=system, y=sf100_c50_ms)
sns.boxplot(ax=axes[0, 2], x=system, y=sf1_c100_ms)
sns.boxplot(ax=axes[1, 0], x=system, y=sf1_c50_tps)
sns.boxplot(ax=axes[1, 1], x=system, y=sf100_c50_tps)
sns.boxplot(ax=axes[1, 2], x=system, y=sf1_c100_tps)


```
